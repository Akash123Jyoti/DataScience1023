```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF value'] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
```

# Modelling

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,ExtraTreesRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score,KFold
```

```python
# LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr_pred = lr.predict(x_test)
print('R2_score : ', r2_score(y_test,lr_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,lr_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,lr_pred))
```

```python
# Lasso
l = Lasso(alpha=1.0)
l.fit(x_train,y_train)
l_pred = l.predict(x_test)
print('R2_score : ', r2_score(y_test,l_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,l_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,l_pred))
```

```python
# RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
r_pred = rfr.predict(x_test)
print('R2_score : ', r2_score(y_test,r_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,r_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,r_pred))
```

```python
# GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr.fit(x_train,y_train)
g_pred = gbr.predict(x_test)
print('R2_score : ', r2_score(y_test,g_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,g_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,g_pred))
```

```python
# ExtraTreesRegressor
etr = ExtraTreesRegressor()
etr.fit(x_train,y_train)
e_pred = etr.predict(x_test)
print('R2_score : ', r2_score(y_test,e_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,e_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,e_pred))
```

```python
# Ridge
r = Ridge()
r.fit(x_train,y_train)
rr_pred = r.predict(x_test)
print('R2_score : ', r2_score(y_test,rr_pred))
print('Mean_Squared_Error : ', mean_squared_error(y_test,rr_pred))
print('Mean_Absolute_Error : ', mean_absolute_error(y_test,rr_pred))
```

# Cross Validation

```python
# LinearRegression
k_folds = KFold(n_splits=5)
score = cross_val_score(lr,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,lr_pred) - score.mean()
```

```python
# Lasso
k_folds = KFold(n_splits=5)
score = cross_val_score(l,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,l_pred) - score.mean()
```

```python
# RandomForestRegressor
k_folds = KFold(n_splits=5)
score = cross_val_score(rfr,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,r_pred) - score.mean()
```

```python
# GradientBoostingRegressor
k_folds = KFold(n_splits=5)
score = cross_val_score(gbr,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,g_pred) - score.mean()
```

```python
# ExtraTreesRegressor
k_folds = KFold(n_splits=5)
score = cross_val_score(etr,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,e_pred) - score.mean()
```

```python
# Ridge
k_folds = KFold(n_splits=5)
score = cross_val_score(r,x,y,cv=k_folds)
print("Cross Validation Score : {}".format(score))
print("Average Cross Validation Score : {}".format(score.mean()))
print("\n------------------")
r2_score(y_test,rr_pred) - score.mean()
```

# Here Three models like Linear Regression, Lasso , Ridge regression are our best model as the difference between r2_score and cross validation score is least.We can use one of them.
